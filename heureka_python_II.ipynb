```python
import requests
from bs4 import BeautifulSoup
```

```python
hledam = input("Zadej co chceš hledat: ")
spravne_zadane = hledam.replace(" ","+")
url = f"https://www.heureka.cz/?h%5Bfraze%5D={spravne_zadane}"
print(url)
```

```python
r = requests.get(url)
dir(r)
```

```python
html = r.text
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
help(soup)
```

```python
for elemen in soup.find_all("a"):
  try:
    if elemen["href"].startswith("https"):
      print(elemen["href"])
  except KeyError as ve:
    print("Nema atribut href")
```

```python
h_letters = [elenement['href'] for elenement in soup.select("h3 a ")]
print(h_letters)
```

```python
list_url_adress = []
for element in soup.select("h3 a"):
  list_url_adress.append(element["href"])
list_url_adress
```

```python
url1 = list_url_adress[0]
html = requests.get(url1).text
soup = BeautifulSoup(html)
for element in soup.select("h1"):
  print(element.text)
```

```python
for element in soup.select("span.u-gamma"):
  print(element.text.replace(" ", "").replace("\n",""))
```
